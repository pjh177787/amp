```python
def test(depth, expanded):
    if depth == 0:
        for i in range(10):
            print(expanded)
            expanded += 1
        return expanded
    if depth%2 == 0:
        expanded_local = 0
        for i in range(10):
            r_expanded = test(depth - 1, expanded)
            print(expanded)
            expanded_local += r_expanded
    else:
        expanded_local = 0
        for i in range(10):
            r_expanded = test(depth - 1, expanded)
            print(expanded)
            expanded_local += r_expanded
    return expanded_local + 1

print(test(3, 0))
```
