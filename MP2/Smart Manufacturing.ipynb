## 1.1 Planning Using A* Search

```python
from search import *

def heuristic_step(f, widgets, cost):
    count = 0
    for widget in widgets:
        depth = fmap.get_len(widget)
        for i in range(depth):
            factory = widget.pop()
            if factory == f:
                count += 1
            widget.push(factory)
    return cost + count
        
def heuristic_dist(f, widgets, cost):
    frontier = fmap.get_frontier(widgets)
    total = 0
    count = 0
    for factory in frontier:
        if not factory == f:
            total += fmap.get_dist(f, factory)
            count += 1
    adist = total/count
        
    max_depth = 0
    for widget in widgets:
        depth = fmap.get_len(widget)
        if depth > max_depth:
            max_depth = depth
            
    return cost + adist*max_depth

def heuristic(f, widgets):
    total_count = 0
    count = 0
    total_dist = 0
    for widget in widgets:
        for i in range(fmap.get_len(widget)):
            factory = widget.pop()
            if factory == f:
                count += 1
            else:
                total_dist = fmap.get_dist(f, factory)
            widget.push(factory)
            total_count += 1
#             print(total_dist)
    return total_dist[0]/(total_count - count)

fmap = Factory_Map()
graph = fmap.get_graph()
widget1 = fmap.get_widget1()
widget2 = fmap.get_widget2()
widget3 = fmap.get_widget3()
widget4 = fmap.get_widget4()
widget5 = fmap.get_widget5()

total_path = ''
widgets = [widget1, widget2, widget3, widget4, widget5]
```
