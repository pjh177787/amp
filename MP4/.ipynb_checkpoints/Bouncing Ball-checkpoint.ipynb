```python
import sys, pygame
from time import sleep

pygame.init()
BLACK = 0, 0, 0
WHITE = 255, 255, 255

screen_size = width, height = 720, 450

ball_direction = [-1, 1]
ball_position = [int(width/2), int(height/2)]
ball_size = 10
ball_speed = 2

paddle_width = 3
paddle_height = 50
paddle_rect = [680, height/2, paddle_width, paddle_height]

screen = pygame.display.set_mode(screen_size)
key_count = 0

while 1:
    sleep(0.01/ball_speed)
    # It needs to be slow because the pygame is.
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT: 
            sys.exit()
            
    if ball_position[0] - ball_size/2 < 0 or ball_position[0] + ball_size/2 > width:
        ball_direction[0] = -ball_direction[0]
    if ball_position[1] - ball_size/2 < 0 or ball_position[1] + ball_size/2 > height:
        ball_direction[1] = -ball_direction[1]
    
    ball_position[0] += ball_direction[0]
    ball_position[1] += ball_direction[1]
       
    events = pygame.event.get()
    if event.type == pygame.KEYDOWN:
        if event.key == pygame.K_UP and paddle_rect[1] > 0:
            print('UP')
            key_count += 1
            if key_count == 5:
                paddle_rect[1] -= 5
                print('UP')
                key_count = 0
        if event.key == pygame.K_DOWN and paddle_rect[1] < height - paddle_height:
            print('DOWN')
            key_count += 1
            if key_count == 5:
                paddle_rect[1] += 5
                print('DOWN')
                key_count = 0
    
    if ball_position[0] + ball_size/2 == paddle_rect[0]:
        if ball_position[1] < paddle_rect[1] + paddle_rect[3] + ball_size/2 and ball_position[1] > paddle_rect[1] - ball_size/2 :
            ball_direction[0] = -ball_direction[0]
            
    
    if ball_position[0] - ball_size/2 > paddle_rect[0]:
        for i in range(5):
            screen.fill(WHITE)
            pygame.display.update()
            sleep(0.2)
            screen.fill(BLACK)
            pygame.draw.rect(screen, WHITE, paddle_rect)
            pygame.draw.circle(screen, WHITE, ball_position, ball_size)
            pygame.display.update()
            sleep(0.2)
        break
        
    screen.fill(BLACK)
    pygame.draw.rect(screen, WHITE, paddle_rect)
    pygame.draw.circle(screen, WHITE, ball_position, ball_size)
    pygame.display.update()
```
