```python
import sys, pygame
import numpy as np
from time import sleep
from Pong import *
from Automation import *
from matplotlib import pyplot as plt

c = 10
gamma = 0.9
ne = 100

my_agent = Pong.Agent((12, 12), lambda x: c / (c + x), gamma, ne)
game = Pong(0.5, 0.5, 0.03, 0.01, 0.5 - Pong.PADDLE_HEIGHT / 2, (12, 12))

name = '_'.join(['c'+str(c), 'gamma'+str(gamma), 'ne'+str(ne)])
directory = './training_result/'
training_file = 'training_' + name + '_file.txt'

if find(directory, training_file):
    with open(directory + training_file, 'r') as f:
        training = f.read()
        my_agent.action_utility = parse_utility(decode(training))
else:
    if is_train == 0:
        print("not input file:", directory + training_file)
        
pygame.init()
BLACK = 0, 0, 0
WHITE = 255, 255, 255

screen_size = width, height = 500, 480

ball_direction = [-1, 1]
ball_position = [int(width/2), int(height/2)]
ball_size = 10
ball_speed = 2

paddle_width = 3
paddle_height = 50
paddle_rect = [480, height/2, paddle_width, paddle_height]

screen = pygame.display.set_mode(screen_size)
key_count = 0

while 1:
    sleep(0.1/ball_speed)
    # It needs to be slow because the pygame is.
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT: 
            sys.exit()
    
    curr_state = game.state.get_state()
    ball_x, ball_y, velocity_x, velocity_y, paddle_y = curr_state
    ball_position = [int(ball_x*width/12), int(ball_y*height/12)]
    paddle_rect[1] = int(paddle_y*height/12)
    
    print(ball_x)
    
    if ball_position[0] - ball_size/2 > paddle_rect[0]:
        for i in range(5):
            screen.fill(WHITE)
            pygame.display.update()
            sleep(0.2)
            screen.fill(BLACK)
            pygame.draw.rect(screen, WHITE, paddle_rect)
            pygame.draw.circle(screen, WHITE, ball_position, ball_size)
            pygame.display.update()
            sleep(0.2)
        break
        
    screen.fill(BLACK)
    pygame.draw.rect(screen, WHITE, paddle_rect)
    pygame.draw.circle(screen, WHITE, ball_position, ball_size)
    pygame.display.update()
    
    next_action = my_agent.policy(curr_state)
    game.proceed(next_action)
    next_state = game.state.get_state()
    my_agent.update_utility(curr_state, next_state, next_action, game.state.score)
```

```python
pygame.init()
BLACK = 0, 0, 0
WHITE = 255, 255, 255

screen_size = width, height = 500, 480

ball_direction = [1, 1]
ball_position = [int(width/2), int(height/2)]
ball_size = 10
ball_speed = 2

paddle_width = 3
paddle_height = 50
paddle1_rect = [20, height/2, paddle_width, paddle_height]
paddle2_rect = [480, height/2, paddle_width, paddle_height]

screen = pygame.display.set_mode(screen_size)
key_count = 0

while 1:
    sleep(0.02/ball_speed)
    # It needs to be slow because the pygame is.
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT: 
            sys.exit()
            
    if ball_position[0] - ball_size/2 < 0 or ball_position[0] + ball_size/2 > width:
        ball_direction[0] = -ball_direction[0]
        
    if ball_position[1] - ball_size/2 < 0 or ball_position[1] + ball_size/2 > height:
        ball_direction[1] = -ball_direction[1]
    
    ball_position[0] += ball_direction[0]
    ball_position[1] += ball_direction[1]
       
    events = pygame.event.get()
    if event.type == pygame.KEYDOWN:
        if event.key == pygame.K_UP and paddle1_rect[1] > 0:
            print('UP')
            key_count += 1
            if key_count == 5:
                paddle1_rect[1] -= 5
                print('UP')
                key_count = 0
        if event.key == pygame.K_DOWN and paddle1_rect[1] < height - paddle_height:
            print('DOWN')
            key_count += 1
            if key_count == 5:
                paddle1_rect[1] += 5
                print('DOWN')
                key_count = 0
    
    
    
    
    
    
    if ball_position[0] - ball_size/2 == paddle1_rect[0]:
        if ball_position[1] < paddle1_rect[1] + paddle1_rect[3] + ball_size/2 and ball_position[1] > paddle1_rect[1] - ball_size/2 :
            ball_direction[0] = -ball_direction[0]
    if ball_position[0] - ball_size/2 == paddle2_rect[0]:
        if ball_position[1] < paddle2_rect[1] + paddle2_rect[3] + ball_size/2 and ball_position[1] > paddle2_rect[1] - ball_size/2 :
            ball_direction[0] = -ball_direction[0]
            
    
    if ball_position[0] + ball_size/2 < paddle1_rect[0]:
        for i in range(5):
            screen.fill(WHITE)
            pygame.display.update()
            sleep(0.2)
            screen.fill(BLACK)
            pygame.draw.rect(screen, WHITE, paddle1_rect)
            pygame.draw.rect(screen, WHITE, paddle2_rect)
            pygame.draw.circle(screen, WHITE, ball_position, ball_size)
            pygame.display.update()
            sleep(0.2)
        break
        
    screen.fill(BLACK)
    pygame.draw.rect(screen, WHITE, paddle1_rect)
    pygame.draw.rect(screen, WHITE, paddle2_rect)
    pygame.draw.circle(screen, WHITE, ball_position, ball_size)
    pygame.display.update()
```
